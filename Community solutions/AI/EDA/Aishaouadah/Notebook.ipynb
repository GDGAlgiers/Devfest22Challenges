## Exploratory data analysis 

```python
"""
Exploratory data analysis on the indemnity data
"""
```

```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
## Reading the dataset
df=pd.read_csv("indemnity.csv")
df.head()
```

```python
## Getting general informations
df.info()
```

```python
## Some statistics 
df.describe()
```

```python
## Counting Categorical data
df["chronic illness"].value_counts(normalize=True)
```

```python
## Counting Categorical data
df["area"].value_counts(normalize=True)
```

```python
## trying to see if there's a relationship between age gender and chronic illness
sns.catplot(data=df, y="age", hue="gender", x="chronic illness", kind="bar")
plt.show()
```

```python
df.corr()
```

```python
## cheking relation between all variables (pairs)
sns.pairplot(df)
```

```python

# from the pair plot above body mass index and bills seems having a relation

#BMI Categories:
#Underweight = <18.5
#Normal weight = 18.5-24.9
#Overweight = 25-29.9
#Obesity = BMI of 30 or greater 

sns.scatterplot(data=df,y="bills", x="Body mass index",hue="chronic illness")
plt.show()
```

***people with chronic illness has more bills to pay***           
***people having an overweight and chronic illness haver higher bills to pay***
